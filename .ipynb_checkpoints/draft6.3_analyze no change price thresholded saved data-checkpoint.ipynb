```python
Ttransactions = pd.read_csv("./results/50_traders_no_change_price/Ttransactions_50p_100s_no_pricechange_thresholded.csv")
TstockChars = pd.read_csv("./results/50_traders_no_change_price/TstockCharacteristics_no_pricechange_thresholded.csv")
TstockPool = pd.read_csv("./results/50_traders_no_change_price/TstockPool_100s_no_pricechange_thresholded.txt")
ThurstPool = pd.read_csv("./results/50_traders_no_change_price/ThurstPool_100s_no_pricechange_thresholded.txt")
```

```python
def findPortfCascades(transactions, minAct = 5):
    """
    works on Transactions table
    """
    if len(transactions) > 0:
        ToS = transactions.ToS.unique()
        tmin = min(transactions.ToS)
        tmax = max(transactions.ToS)
        rebalances = np.array([])
        cascadeSizes = np.array([])
        keySep = np.array([])
        for i,t in enumerate(range(tmin,tmax+1)):
            portfs = np.unique(np.append(transactions[transactions.ToS==t].buyer.unique(),
                                         transactions[transactions.ToS==t].seller.unique()))
            rebalance = len(portfs) - 1 # the world account
            rebalances = np.append(rebalances, rebalance)
            if rebalance <= minAct and i != 0:
                keySep = np.append(keySep, int(i))

        i0 = 0
        for key in keySep:
            key += 1
            cascadeSize = np.sum(rebalances[int(i0):int(key)])
            cascadeSizes = np.append(cascadeSizes,cascadeSize)
            i0=key
        return cascadeSizes
    else:
        return []
    
def findStockCascades(stockTrans, maxSep = 10):
    """
    works on stockTransactions tables
    """
    
    if len(stockTrans) > 0 :
        ToS = stockTrans.ToS.unique()
        sep = np.diff(ToS)
        keySep = np.where(sep>maxSep)[0]

        i0 = 0
        cascadeSizes = np.asarray([])
        for key in keySep:
            key += 1
    #         print(i0, key)
            times = ToS[i0:key]
            cascadeSize = len(stockTrans[stockTrans.ToS.isin(times)])
            cascadeSizes = np.append(cascadeSizes,cascadeSize)
            i0 = key

        return cascadeSizes
    else:
        return []
```

```python
totalCascades = np.asarray([])
plt.close()
for s in range(100):
    stockTrans = Ttransactions[Ttransactions['stock']==s]
    if len(stockTrans) > 0:
#         stockTrans.hist(column='ToS', bins=np.arange(992,2000))
#         plt.title("Transactions of stock: "+ str(s))
#         plt.show()
        cascades = findStockCascades(stockTrans, 1)
#         print(s, len(cascades))
        totalCascades = np.append(totalCascades,cascades)
```

```python
print(max(totalCascades))
```

```python
hist, bins = np.histogram(totalCascades, bins=np.logspace(np.log10(1),np.log10(2700),25))
binsize = np.diff(bins)

mid = np.array([])
for i in range(len(binsize)):
    avg = (bins[i]+bins[i+1])/2
    mid = np.append(mid,avg)
```

```python
plt.plot(mid,np.divide(hist,binsize)/len(totalCascades),'o-', linewidth = 1, markersize = 3)
plt.plot(mid,np.power(mid,-1.7)*12, '-.', label= "$x^{-1.7}$")
plt.loglog()
plt.legend()
plt.title("Cascade size distributions (no price change) max time sep : 1")
plt.ylabel('P(S)')
plt.xlabel('S')
plt.grid(True)
# plt.savefig("./results/50_traders_no_change_price/cascadesize_nopc_stockTrans_maxsep_1.png", dpi=250)
plt.show()
plt.close()
```
